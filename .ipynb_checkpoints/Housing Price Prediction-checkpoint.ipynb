```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('https://d24c4belkzmc2t.cloudfront.net/cdn-uploads/1608527992_housing_price.csv')
```

```python
data.head()
```

```python
data.dropna(inplace = True)
```

```python
data.corr().iloc[-1]
```

```python
l1, l2, l3, l4 = LabelEncoder(), LabelEncoder(), LabelEncoder(), LabelEncoder()

data['HouseType'] = l1.fit_transform(data['HouseType'])
data['Foundation'] = l2.fit_transform(data['Foundation'])
data['CentralAC'] = l3.fit_transform(data['CentralAC'])
data['GarageType'] = l4.fit_transform(data['GarageType'])
```

```python
data.head()
```

```python
data.corr().iloc[-1]
```

```python
x = data[['Quality', 'BasementArea (ft2)', 'Area (ft2)', 'BathroomNos', 
          'BedroomsNos', 'GarageCars', 'GarageArea (ft2)']]

y = data['SalePrice']
```

```python
sc = StandardScaler()
```

```python
x_new = sc.fit_transform(x)
```

### Model 1

```python
m1_dic = {}
m1_r = {}

for i in range(1, 400):
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.05, random_state = i)
    model1 = LinearRegression()
    model1.fit(xtrain, ytrain)
    m1_r[i] = model1.score(xtrain, ytrain)
    m1_dic[i] = mean_absolute_error(ytest, model1.predict(xtest))
```

```python
min_value = min(m1_dic.values())
for key, value in m1_dic.items():
    if value == min_value:
        print(key, min_value, m1_r[key])
```

### Model 2 

```python
m2_dic = {}
m2_r = {}

for i in range(1, 200):
    xtrain, xtest, ytrain, ytest = train_test_split(x_new, y, test_size = 0.05, random_state = i)
    model2 = LinearRegression()
    model2.fit(xtrain, ytrain)
    m2_r[i] = model2.score(xtrain, ytrain)
    m2_dic[i] = mean_absolute_error(ytest, model2.predict(xtest))
```

```python
min_value = min(m2_dic.values())
for key, value in m2_dic.items():
    if value == min_value:
        print(key, min_value, m2_r[key])
```
