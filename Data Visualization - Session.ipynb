## Changing size of a figure and plotting two points connected by a line using plot function

```python
plt.figure(figsize=(15, 10)) # expressed in inches
plt.plot([1, 2],[2, 3])
plt.show()
```

```python
''' The format of plot function is 

plt.plot(values of x, values of y, formatting)
    - values of x = x values in form of list/numpy arrays/pandas column
    - values of y = y values in form of list/numpy arrays/pandas column
    - formatting = formatting is in this order
            color, marker type, type of line
                    --> Possible colours are red/blue/green
                    --> Possible markers are + * o
                    --> Possible type of lines are - (for straight line), -- (for dashed line)
                    
'''
```

```python
plt.figure(figsize=(15, 10)) 
plt.plot([1, 2],[2, 3], 'b+-')
plt.show()
```

```python
# Changing line type to dashed type

plt.figure(figsize=(15, 10)) 
plt.plot([1, 2],[2, 3], 'b+--')
plt.show()
```

```python
# Changing markers to circle

plt.figure(figsize=(15, 10)) 
plt.plot([1, 2],[2, 3], 'b+-')
plt.show()
```

## Legend in matplotlib

```python
x = np.arange(-8, 8)
x
```

```python
y_cos = np.cos(x)
```

```python
y = np.sin(x)
```

```python
plt.figure(figsize = (10, 6))
plt.plot(x, y) # sine curve
plt.plot(x, y_cos) # cos curve
plt.title('Sin Cos Graph')
plt.xlabel('X variable')
plt.ylabel('Plottled Values')
plt.legend(['Sin', 'Cos'])
plt.xlim(-8,8)
plt.ylim(-1,1)
plt.show()
```

## Limiting x and y axis values using xlim and ylim function respectively

```python
x = np.arange(-100, 100)
y = np.sin(x)
plt.plot(x, y, 'r')
plt.xlim(-25, 25)
plt.ylim(-0.25, 0.25)
plt.show()
```

## Line Plot

### Growth of coronavirus cases in India on a daily basis

```python
data = pd.read_csv('https://d24c4belkzmc2t.cloudfront.net/cdn-uploads/1590748694_Covid_Dataset.csv')
data.head(10)
```

```python
india_data = data[data['Country/Region'] == 'India']
```

```python
india_data['Date'] = pd.to_datetime(india_data['Date'])
```

```python
india_data['Date'].iloc[0]
```

```python
x = india_data['Date']
y = india_data['Confirmed']

fig = plt.figure(figsize = (15, 10))
plt.plot(x, y, 'b+')
plt.title('Number of cases in India on daily basis')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.show()
```

## Bar Graph

### Top 10 most infected countries in the world

```python
# groupby

country_wise_data = data.groupby('Country/Region').sum()
country_wise_data.head()
```

```python
sorted_data = country_wise_data.sort_values(by = 'Confirmed', ascending = False).head(10)
sorted_data
```

```python
fig = plt.figure(figsize = (15, 10))
plt.bar(sorted_data.index, sorted_data['Confirmed'])
plt.show()
```

## Scatter plot

## Histograms

### Histograms for number of confirmed cases with no bins defined (matplotlib automatically find out bins)

```python
plt.figure(figsize=(15,9))
plt.hist(country_wise_data['Confirmed'])
plt.xlabel('No of Confirmed Cases')
plt.ylabel('No. of Countries')
plt.title('Relation Between the range of confirmed cases vs The no. of Countries')
```

### Histograms for number of confirmed cases with bins defined

```python
fig = plt.figure(figsize = (10, 8))

plt.hist(country_wise_data['Confirmed'],
         bins = np.arange(0, 30000000, 1000000))
plt.show()
```

## Kdeplot

### Kde plot for confirmed cases

```python
fig = plt.figure(figsize = (10, 8))

sns.kdeplot(country_wise_data['Confirmed'])
plt.show()
```

### 2 KDE plots of Deaths and Confirmed in single figure with different colours

```python
fig = plt.figure(figsize = (10, 8))

sns.kdeplot(country_wise_data['Deaths'], color = 'orange')
sns.kdeplot(country_wise_data['Confirmed'], color = 'blue')
plt.show()
```

**Observation**: When confirmed cases are more than deaths are also more and when cases started decreasing then deaths decreases not exponentially but gradually as previous positive cases can also be converted into casualities.

## Lmplot
